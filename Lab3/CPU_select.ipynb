# CPU_select

```python
import subprocess

ncpu = int(input("Which cpu to use? "))
if ncpu<0 or ncpu>1:
    print("0 or 1 only.")
else:
    nterms = input("How many terms? ")

    subprocess.run(
        ["taskset", "-c", str(ncpu), "python3", "fib.py"],
        input=nterms + "\n",
        text=True
    )
```
